```python
def crear_spacetime_cube(df, paises_seleccionados=None, top_n_puertos=20, separar_cercanos=True, tama√±o_base=5, zoom_geografico=1.0):
    """
    Crea Space-Time Cube interactivo con normalizaci√≥n de coordenadas
    
    Par√°metros:
    - zoom_geografico: Factor de zoom (1.0 = normal, 2.0 = m√°s separado, 0.5 = m√°s compacto)
    """
    
    if paises_seleccionados:
        df_viz = df[df['Pa√≠s'].isin(paises_seleccionados)].copy()
    else:
        df_viz = df.copy()
    
    if len(df_viz) == 0:
        print("‚ö†Ô∏è No hay datos para los pa√≠ses seleccionados")
        return None
    
    # Top puertos
    top_puertos = df_viz.groupby('Puerto')['Valor US$'].sum().nlargest(top_n_puertos).index
    df_viz = df_viz[df_viz['Puerto'].isin(top_puertos)]
    
    # MEJORA CLAVE: Normalizar coordenadas geogr√°ficas a escala 0-100
    # Esto hace que la separaci√≥n visual sea m√°s clara
    lat_min, lat_max = df_viz['lat'].min(), df_viz['lat'].max()
    lon_min, lon_max = df_viz['lon'].min(), df_viz['lon'].max()
    
    # Normalizar a escala 0-100 (m√°s f√°cil de trabajar)
    df_viz['lat_norm'] = ((df_viz['lat'] - lat_min) / (lat_max - lat_min)) * 100
    df_viz['lon_norm'] = ((df_viz['lon'] - lon_min) / (lon_max - lon_min)) * 100
    
    # Separar puertos cercanos en el espacio NORMALIZADO
    if separar_cercanos:
        # Offsets ahora en escala normalizada (m√°s efectivos visualmente)
        offsets_puertos = {
            # Los Angeles Area - Separaci√≥n amplia en escala normalizada
            'LONG BEACH, CA': {'lat_offset': -3, 'lon_offset': 4},
            'LOS ANGELES, CA': {'lat_offset': 0, 'lon_offset': 0},
            'LOS ANGELES INTERNATIONAL AIRPORT, CA': {'lat_offset': 3, 'lon_offset': -4},
            
            # Seattle/Tacoma Area
            'SEATTLE, WA': {'lat_offset': 2.5, 'lon_offset': 2.5},
            'TACOMA, WA': {'lat_offset': -2.5, 'lon_offset': -2.5},
            'SEATTLE-TACOMA INTERNATIONAL AIRPORT, WA': {'lat_offset': 0, 'lon_offset': 5},
            
            # New York Area
            'JFK INTERNATIONAL AIRPORT, NY': {'lat_offset': 2.5, 'lon_offset': 3},
            'NEW YORK, NY': {'lat_offset': 0, 'lon_offset': 0},
            'NEWARK, NJ': {'lat_offset': -2.5, 'lon_offset': -3},
            
            # San Francisco Bay Area
            'OAKLAND, CA': {'lat_offset': 2.5, 'lon_offset': 2.5},
            'SAN FRANCISCO, CA': {'lat_offset': -2.5, 'lon_offset': -2.5},
            'SAN FRANCISCO INTERNATIONAL AIRPORT, CA': {'lat_offset': 0, 'lon_offset': 5},
            
            # Houston Area
            'HOUSTON, TX': {'lat_offset': 0, 'lon_offset': 0},
            'HOUSTON INTERCONTINENTAL AIRPORT, TX': {'lat_offset': 2.5, 'lon_offset': 3},
            
            # Miami Area
            'MIAMI, FL': {'lat_offset': 0, 'lon_offset': 0},
            'MIAMI INTERNATIONAL AIRPORT, FL': {'lat_offset': 2.5, 'lon_offset': 3},
            'PORT EVERGLADES, FL': {'lat_offset': -2.5, 'lon_offset': -2.5},
            
            # Philadelphia Area
            'PHILADELPHIA, PA': {'lat_offset': 0, 'lon_offset': 0},
            'PHILADELPHIA INTERNATIONAL AIRPORT, PA': {'lat_offset': 2.5, 'lon_offset': 3},
            
            # Baltimore/DC Area
            'BALTIMORE, MD': {'lat_offset': 0, 'lon_offset': 0},
            'BALTIMORE-WASHINGTON INTERNATIONAL AIRPORT, MD': {'lat_offset': 2.5, 'lon_offset': 3},
            'WASHINGTON, DC': {'lat_offset': -2.5, 'lon_offset': -2.5},
            
            # Chicago Area
            'CHICAGO, IL': {'lat_offset': 0, 'lon_offset': 0},
            'CHICAGO EXECUTIVE AIRPORT, IL (FORMALLY PAL-WAUKEE': {'lat_offset': 2.5, 'lon_offset': 3},
        }
        
        # Aplicar offsets en escala normalizada
        df_viz['lat_viz'] = df_viz.apply(
            lambda row: row['lat_norm'] + offsets_puertos.get(row['Puerto'], {}).get('lat_offset', 0) * zoom_geografico,
            axis=1
        )
        df_viz['lon_viz'] = df_viz.apply(
            lambda row: row['lon_norm'] + offsets_puertos.get(row['Puerto'], {}).get('lon_offset', 0) * zoom_geografico,
            axis=1
        )
    else:
        df_viz['lat_viz'] = df_viz['lat_norm']
        df_viz['lon_viz'] = df_viz['lon_norm']
    
    # Normalizar tiempo
    a√±o_min = df_viz['A√±o'].min()
    a√±o_max = df_viz['A√±o'].max()
    df_viz['time_normalized'] = (df_viz['A√±o'] - a√±o_min) / (a√±o_max - a√±o_min)
    
    # Tama√±o de c√≠rculos
    df_viz['size'] = np.log10(df_viz['Valor US$'] + 1) * tama√±o_base
    
    print(f"üìä Visualizando: {len(df_viz):,} registros | {df_viz['Pa√≠s'].nunique()} pa√≠ses | {df_viz['Puerto'].nunique()} puertos")
    print(f"üìç Rango geogr√°fico: Lat {lat_min:.1f}¬∞ a {lat_max:.1f}¬∞ | Lon {lon_min:.1f}¬∞ a {lon_max:.1f}¬∞")
    
    # Crear figura
    fig = go.Figure()
    
    for pais in sorted(df_viz['Pa√≠s'].unique()):
        df_pais = df_viz[df_viz['Pa√≠s'] == pais]
        
        fig.add_trace(go.Scatter3d(
            x=df_pais['lon_viz'],
            y=df_pais['lat_viz'],
            z=df_pais['time_normalized'],
            mode='markers',
            name=pais,
            marker=dict(
                size=df_pais['size'],
                color=df_pais['Valor US$'],
                colorscale='Viridis',
                showscale=True,
                colorbar=dict(
                    title="Valor US$",
                    x=1.1,
                    tickformat='$,.0f'
                ),
                line=dict(width=0.5, color='white'),
                opacity=0.85
            ),
            text=[f"<b>{puerto}</b><br>Pa√≠s: {pais}<br>A√±o: {a√±o}<br>Valor: ${valor:,.0f}<br>Ubicaci√≥n: ({lat:.2f}¬∞, {lon:.2f}¬∞)"
                  for puerto, a√±o, valor, lat, lon in zip(df_pais['Puerto'], df_pais['A√±o'], df_pais['Valor US$'], df_pais['lat'], df_pais['lon'])],
            hovertemplate='%{text}<extra></extra>'
        ))
    
    # Layout con Pac√≠fico - Atl√°ntico
    fig.update_layout(
    title=dict(
        text=f'<b>Space-Time Cube: Comercio USA ({a√±o_min}-{a√±o_max})</b>',
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    scene=dict(
        xaxis=dict(
            title='<b>Pac√≠fico - Atl√°ntico</b>',
            showgrid=True,
            gridcolor='lightgray',
            range=[0, 100]
        ),
        yaxis=dict(
            title='<b>Sur - Norte</b>',
            showgrid=True,
            gridcolor='lightgray',
            range=[0, 100]
        ),
        zaxis=dict(
            title='<b>Tiempo</b>',
            tickvals=np.linspace(0, 1, a√±o_max - a√±o_min + 1),
            ticktext=[str(y) for y in range(a√±o_min, a√±o_max + 1)],
            showgrid=True,
            gridcolor='lightgray'
        ),
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.3)),
        bgcolor='rgba(240, 240, 240, 0.9)',
        aspectmode='cube'
    ),
    width=1200,
    height=800,
    showlegend=True,
    legend=dict(
        font=dict(size=11),
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.01
    )
)
    
    return fig
```
